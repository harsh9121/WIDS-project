```python
from scipy.io import loadmat
mnist = loadmat(r"C:\Users\sivar\Downloads\mnist-original.mat\mnist-original.mat")
images = mnist["data"].T
labels = mnist["label"][0]
label_counts = [0] * 10  # Initialize a list to store counts for each digit

for i in labels:
    label_counts[int(i)] += 1

# Display the distribution of labels
plt.figure(figsize=(8, 5))
plt.bar(range(10), label_counts, color='skyblue')
plt.xlabel('Digit')
plt.ylabel('Frequency')
plt.title('Distribution of Labels (Digits 0-9)')
plt.xticks(range(10))
plt.grid(axis='y')
plt.show()

```
